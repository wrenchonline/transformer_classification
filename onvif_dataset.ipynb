```python
#coding:utf-8
import sys
import os
import lxml
import glob
import shutil
from bs4 import BeautifulSoup
import unicodedata
import re

Dahua = 1
i = 0
j = 0
strd = ""

if Dahua:
    i = 122
    j = 129 
    strd = "0" 
else:
    i = 2
    j = 80 
    strd = "1"
def unicode_to_ascii(s):
    return "".join(
        c for c in unicodedata.normalize("NFD", s) if unicodedata.category(c) != "Mn"
    )

def preprocess_sentence(w):

    w = unicode_to_ascii(w.lower().strip())

    string = re.sub(r"[^A-Za-z0-9(),!?\'\`]", " ", w)
    string = re.sub(r"\'s", " \'s", string)
    string = re.sub(r"\'ve", " \'ve", string)
    string = re.sub(r"n\'t", " n\'t", string)
    string = re.sub(r"\'re", " \'re", string)
    string = re.sub(r"\'d", " \'d", string)
    string = re.sub(r"\'ll", " \'ll", string)
    string = re.sub(r",", " , ", string)
    string = re.sub(r"!", " ! ", string)
    string = re.sub(r"\(", " \( ", string)
    string = re.sub(r"\)", " \) ", string)
    string = re.sub(r"\?", " \? ", string)
    string = re.sub(r"\s{2,}", " ", string)
    return string.strip().lower()

root_dir = '/home/wrench/下载/ML_Tech_Stack/Natural_language_process/'
parent_dir = '/home/wrench/下载/ML_Tech_Stack/Natural_language_process/data_onvif/1/'
new_tran = '/home/wrench/下载/ML_Tech_Stack/Natural_language_process/data_onvif/2/val.txt'
HKfilenames =  []

if Dahua:
    s = 11
    z = 17
    for iz in range(s,z):
        for index in  range(i,j):
            filename = '172.30.' + str(iz) + '.' + str(index) + "_result.xml"
            fsize = os.path.getsize(os.path.join(parent_dir, filename))
            if not fsize:
                continue
            HKfilenames.append(os.path.join(parent_dir, filename))
else:
    for index in  range(i,j):
        filename = '172.30.11.' + str(index) + "_result.xml"
        fsize = os.path.getsize(os.path.join(parent_dir, filename))
        if not fsize:
            continue
        HKfilenames.append(os.path.join(parent_dir, filename))
    
    
result_list = []
for index,xml_file in enumerate(HKfilenames):
    with open(xml_file, 'r+',encoding='utf-8') as f:
        xml = f.read()
        soup = BeautifulSoup(xml, 'xml')
        result = soup.find('d:Scopes')
        for r in result:
            r = preprocess_sentence(r)
            print(r)
            result_list.append(r + "\t" + strd + "\n")
        print('-'*20)
with open(new_tran, 'a+',encoding='utf-8') as f1:
      for result in result_list:
            f1.write(result)
```
